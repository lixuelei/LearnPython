```python
import random, string
# 创建一个tag最好的方法是调用工厂方法 BeautifulSoup.new_tag() :
soup = BeautifulSoup("<b></b>", 'lxml')
original_tag = soup.b
new_tag = soup.new_tag('Attribute', Attribute_1 = random.sample(string.ascii_uppercase, 1))
original_tag.append(new_tag)
original_tag
```

```python
from bs4 import BeautifulSoup
import random, string
soup = BeautifulSoup(open('running-example.xes'), "lxml")

num = random.randint(5,10)

for event in soup.find_all('event'):
    for i in range(1, num):
        new_tag = soup.new_tag('string', key = 'Attribute_%s' % (i), value = random.sample(string.ascii_uppercase, 1))
        event.append(new_tag)
    print(event.prettify())
    with open('save.xes', 'w') as f:
        f.write(str(soup))
```
